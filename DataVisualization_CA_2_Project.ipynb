# **Univariate Analysis**

```python
per=data['percentile'].round()
mean=per.mean()
median=per.median()
mode=per.mode()
std=per.std()
print(f"Mean = {mean}")
print(f"Median = {median}")
print(f"Mode = {mode[0]}")
print(f"Standard deviation = {std}")
#First 50 percentiles
plt.figure(figsize=(10,8))
sns.countplot(x=per[:50])
plt.title("Percentiles Count Plot",color="#FF1493")
plt.xlabel("Percentile",color="#FFD700")
plt.ylabel("Count",color="#FF6A6A")
plt.style.use("dark_background")
#Last 50 percentiles
plt.figure(figsize=(10,8))
sns.countplot(x=per[1160:])
plt.title("Percentiles Count Plot",color="#FF1493")
plt.xlabel("Percentile",color="#FFD700")
plt.ylabel("Count",color="#FF6A6A")
plt.style.use("dark_background")
plt.show()


plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(data['percentile'],bins=20,edgecolor='#030303',color=['#CAFF70'])
plt.title("Percentile",color='#FF1493')
plt.xlabel("Percentile Range",color='#00EEEE')
plt.ylabel("Frequency",color='#BF3EFF')
plt.show()
```

```python
per=data['college_name']
plt.figure(figsize=(14,6))
sns.countplot(x=per[:800])
plt.title("College Name Plot of 800 Students",color='#FF1493')
plt.xlabel("College Name",color='#00EEEE')
plt.ylabel("Count",color='#BF3EFF')
plt.style.use("dark_background")
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.countplot(x=per[800:])
plt.title("College Name Plot of 400 Students",color='#FF1493')
plt.xlabel("College Name",color='#00EEEE')
plt.ylabel("Count",color='#BF3EFF')
plt.style.use("dark_background")
```

# **Multivariate Analysis**

```python
plt.figure(figsize=(14,8))
sns.pairplot(data=data, vars=['rank', 'percentile'], hue='gender', markers=['o', 's'], palette='Set1')
plt.suptitle('Scatterplot Matrix for Rank and Percentile by Gender',color='#FF82AB',alpha=0.8,y=1.02,x=0.6)
plt.show()
```

**Conclusion : -**


*   The scatterplot matrix displays a strong negative correlation between 'rank' and 'percentile,' indicating that as rank increases (lower rank), percentile decreases for both male and female students.



*   This relationship aligns with the typical behavior in competitive exams, where lower ranks signify higher achievement.



*   The scatterplot matrix effectively illustrates the consistency of this inverse relationship between rank and percentile across genders in the MHT-CET dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a violin plot for percentile by gender and category
plt.figure(figsize=(20, 6))
sns.violinplot(data=data, x='category', y='percentile', hue='gender', palette='Set3')
plt.xlabel('Gender',color='#00BFFF')
plt.ylabel('Percentile',color='#FFE4B5')
plt.title('Violin Plot of Percentile by Gender and Category')
plt.xticks(rotation=90,color='#00BFFF')
plt.yticks(color='#FFE4B5')
plt.show()
```

**Conclusion : -**
Percentile distributions vary by category and gender, with some categories showing wider variability and outliers.
Overall, there is a similarity in percentile spreads between males and females.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a boxplot for rank by category
plt.figure(figsize=(15, 6))
sns.boxplot(data=data, x='category', y='rank', palette='Set2',hue='gender')
plt.xlabel('Category',color='#00BFFF')
plt.ylabel('Rank',color='#FFB6C1')
plt.title('Boxplot of Rank by Category',color='m')
plt.xticks(rotation=90,color='#00BFFF')
plt.yticks(color='#FFB6C1')
plt.show()
```

**Conclusion : -**
The boxplot displays the distribution of ranks across different categories and distinguishes between male and female students within each category.
Categories like 'SC' and 'ST' exhibit wider interquartile ranges (IQRs), indicating more variability in ranks.
